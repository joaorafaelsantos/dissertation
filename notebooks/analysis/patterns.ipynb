```
def get_days_since_creation(date):
    return (date - creation_date).days

fig, ax = plt.subplots(figsize=(30,10), dpi=100)

for pattern in patterns:
    if patterns[pattern]:
        date = datetime.strptime(patterns[pattern], "%Y-%m-%d")
        plt.scatter(pattern, get_days_since_creation(date), label=pattern.replace("_", " ").title())
        ax.annotate(pattern, (pattern, get_days_since_creation(date)))

ax.set_xticklabels([])

# ax.legend()
plt.show()
```
