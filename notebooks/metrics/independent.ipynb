```
import matplotlib.pyplot as plt
import numpy as np
import json
import os
```

```
directory = '../../data/final'
for file_name in os.listdir(directory):
    if file_name.endswith(".json"):
        with open(f'{directory}/{file_name}') as json_file:
            name = file_name[:-5]
            print(name)
            
            figures_directory = f"../figures/{name}"
            if not os.path.exists(figures_directory):
                os.makedirs(figures_directory)
            
            data = json.load(json_file)

            stars = np.array([d['stargazers'] for d in data])
            forks = np.array([d['forks'] for d in data])
            issues = np.array([d['open_issues'] for d in data])
            size = np.array([d['size'] for d in data])
            dates = np.array([d['date'] for d in data])

            #stars
            fig, ax = plt.subplots(figsize=(8,4), dpi= 100)
            fig.suptitle(f'Stars - {name}')
            plt.xticks(np.arange(0, len(dates)+1, 5))
            plt.plot(dates, stars)
            # plt.savefig(f"{figures_directory}/stars.png", facecolor='white', transparent=False)
            plt.show()

            #forks
            fig=plt.figure(figsize=(8,4), dpi= 100)
            fig.suptitle(f'Forks - {name}')
            plt.xticks(np.arange(0, len(dates)+1, 5))
            plt.plot(dates, forks)
            # plt.savefig(f"{figures_directory}/forks.png", facecolor='white', transparent=False)
            plt.show()

            #issues
            fig=plt.figure(figsize=(8,4), dpi= 100)
            fig.suptitle(f'Issues - {name}')
            plt.xticks(np.arange(0, len(dates)+1, 5))
            plt.plot(dates, issues)
            # plt.savefig(f"{figures_directory}/issues.png", facecolor='white', transparent=False)
            plt.show()

            #size
            fig=plt.figure(figsize=(8,4), dpi= 100)
            fig.suptitle(f'Size - {name}')
            plt.xticks(np.arange(0, len(dates)+1, 5))
            plt.plot(dates, size)
            # plt.savefig(f"{figures_directory}/size.png", facecolor='white', transparent=False)
            plt.show()
```
